```python
# Import the pandas library

import pandas as pd
pd.options.display.max_rows = 10
```

```python
# Read in the data

crimes_df = pd.read_csv("crime_data.csv")

crimes_df
```

```python
# Drop crimes not from 2016

crimes_df = crimes_df[crimes_df["Date Occurred"].str[-4 : ] == "2016"]

crimes_df
```

```python
from datetime import datetime
import calendar
```

```python
# Find the day of the week the crime occurred

def date_to_datetime(date):
    month, day, year = date.split("/")
    return datetime(int(year), int(month), int(day))

def date_to_weekday(date):
    weekday_num = date_to_datetime(date).weekday()
    return calendar.day_name[weekday_num]

crimes_df["Day Occurred"] = crimes_df["Date Occurred"].apply(date_to_weekday)

crimes_df
```

```python
# Calculate the days until the crime was reported

def days_to_report(dates):
    reported, occurred = dates
    reported_dt = date_to_datetime(reported)
    occurred_dt = date_to_datetime(occurred)
    return (reported_dt - occurred_dt).days

crimes_df["Days To Report"] = crimes_df[
    ["Date Reported", "Date Occurred"]
].apply(days_to_report, axis=1)
```

```python
# Calculates the day of the year (0-365) that the crime took place

def day_of_year(date):
    month, day, year = date.split("/")
    from_date = datetime(2016, 1, 1)
    return (datetime(int(year), int(month), int(day)) - from_date).days

crimes_df["Day of Year"] = crimes_df["Date Occurred"].apply(day_of_year)

crimes_df
```

```python
# Pulls the sunrise and sunset times for 2016 from sunrise-sunset.org

import requests
from time import sleep
from datetime import timedelta
from pytz import timezone
import pytz

def UTC_to_PST(time_str):
    date, time_and_zone = time_str.split("T")
    year, month, day = date.split("-")
    time_of_day, time_zone = time_and_zone.split("+")
    hour, minute, second = time_of_day.split(":")
    return int(datetime(int(year), int(month), int(day), 
                        int(hour), int(minute), 
                        tzinfo=pytz.utc).
               astimezone(timezone("US/Pacific")).
               strftime("%H%M"))
    
LAlat = 34.073851
LAlong = -118.242147

rows = []
curr_day = datetime(2016, 1, 1)

while curr_day <= datetime(2016, 12, 31):
    
    resp = requests.get(
        "https://api.sunrise-sunset.org/json?lat={0}&lng={1}&date={2}&formatted=0".format(
            LAlat, LAlong, curr_day.strftime("%Y-%m-%d")))
    resp_json = resp.json()
    
    sunrise = resp_json["results"]["sunrise"]
    sunset = resp_json["results"]["sunset"]
    
    rows.append({
        "date" : curr_day.strftime("%m/%d/%Y"),
        "sunrise" : UTC_to_PST(sunrise),
        "sunset" : UTC_to_PST(sunset)
    })
    
    curr_day = curr_day + timedelta(days=1)
    
    sleep(0.05)
    
sunset_df = pd.DataFrame(rows).set_index("date")

sunset_df
```

```python
# Determine whether the crime occurred at night or during the day

def time_of_day(date_time):
    
    date, time = date_time
    sunset_day = sunset_df.loc[date]
    sunrise = sunset_day.loc["sunrise"]
    sunset = sunset_day.loc["sunset"]
    
    time_of_day = "NIGHT"
    if sunrise <= time <= sunset:
        time_of_day = "DAY"
    return time_of_day

crimes_df["Time of Day"] = crimes_df[
    ["Date Occurred", "Time Occurred"]
].apply(time_of_day, axis=1)

crimes_df
```

```python
# Convert the Location column into a column with latitude and a column with longitude

location_df = crimes_df["Location "].str.replace(r"[^0-9-,.]", "").str.split(",", 1, expand=True)

crimes_df["Location Lat"] = location_df[0].astype(float)
crimes_df["Location Lng"] = location_df[1].astype(float)

crimes_df.drop("Location ", axis=1, inplace=True)

crimes_df
```

```python
import numpy as np

# Replace the age with NaN if sex is unknown

def convert_age_nan(age_sex):
    
    age, sex = age_sex
    
    # Indicates sex is NaN
    if sex == "X":
        age = np.nan
    
    return age

# Takes the victim sex to ensure that all age = 0.0 aren't changed
crimes_df["Victim Age"] = crimes_df[
    ["Victim Age", "Victim Sex"]
].apply(convert_age_nan, axis=1)

crimes_df
```

```python
# Drop all NaN observations
# Drop the unnecessary

crimes_df = crimes_df[
    ["Time Occurred", "Crime Code", "Victim Age", "Day Occurred", "Days To Report", 
     "Day of Year", "Time of Day", "Location Lat", "Location Lng"]
]

crimes_df.dropna(inplace=True)

crimes_df
```

```python
# Write the output to a new file

with open("crime_data_edited.csv", "w") as f:
    f.write(crimes_df.to_csv(index=False))
    
with open("sunset_data.csv", "w") as f:
    f.write(sunset_df.to_csv(index=True))
```
