```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
crimes_df = pd.read_csv('crime_data_edited.csv')
crimes_df.dropna(inplace=True)
crimes_df.head()
```

```python
len(crimes_df)
```

**Look at distribution of crime codes (lower crime code is more severe)**

There are a large amount of crimes reported that fall between 200 and 300 on the crime code scale, which is relatively low and therefore relatively severe. There are also a lot of crimes reported that fall between 600 and 700, which is a pretty unsever crime code range.

```python
crimes_df['Crime Code'].plot.hist();
```

**Compare crime code (crime severity) with time of day (night or day and actual time on a 0-24:00 scale)**

We can see that time of day does not have an effect on the severity of crimes committed. There is no evidence from these visualizations that more severe crimes occur at night or vice versa. We can see however that the mean time occurred varies a lot for less severe crimes.

```python
times = crimes_df['Time of Day'].map({
    'DAY': 'yellow',
    'NIGHT': 'royalblue'
})
crimes_df.plot.scatter(x='Crime Code', y='Time Occurred', c=times, alpha=0.2);
```

```python
crimes_df.plot.scatter(x='Crime Code', y='Time Occurred', c='skyblue', alpha=0.2);
```

```python
# Specifies a 1 x 2 grid of plots, figsize in inches
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

crimes_day = crimes_df["Time of Day"].unique()
for ax, cri_day in zip(axes, crimes_day):
    crime_type = crimes_df[crimes_df["Time of Day"] == cri_day]
    crime_type['Crime Code'].plot.hist(color='darkturquoise', ax=ax)
    ax.set_title(cri_day)
```

```python
crimes_df.groupby('Crime Code')['Time Occurred'].mean().plot.line().set_ylabel('Average Time of Day');
```

**Crime code (crime severity) compared with victim age**

There doesn't seem to be a relationship with severity of crime and the victim's age. There are in general more crimes reported for victims aged 20-40 years old.

```python
crimes_df.plot.scatter(x='Crime Code', y='Victim Age', c='seagreen', alpha=0.5);
```

```python
crimes_df.groupby('Crime Code')['Victim Age'].mean().plot.line().set_ylabel('Mean Victim Age');
```

```python
crimes_df['Victim Age'].plot.hist();
```

**Day of the week for crime code**

While there are most crimes occurring on Sunday, and least on Tuesday-Thursday, the amount of crimes each day doesn't differ greatly. The severity of crimes also does not change greatly day by day. 

```python
# Specifies a 1 x 7 grid of plots, figsize in inches
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 7, figsize=(20, 4), sharex=True, sharey=True)

thedays = crimes_df["Day Occurred"].unique()
for ax, cri_day in zip(axes, thedays):
    crime_type = crimes_df[crimes_df["Day Occurred"] == cri_day]
    crime_type['Crime Code'].plot.hist(color='darkturquoise', ax=ax)
    ax.set_title(cri_day)
    ax.set_xlabel('Crime Code')
#     plt.tight_layout()
```

```python
crimes_df['Day Occurred'].value_counts(
    )[['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']].plot.bar(color='orangered');
```

**Looking at Crime Severity, vicitm age, and days to report**

There's not a lot of correlation between days to report and crime severity but some of the less severe crimes have unusually high average days to report. Victim age has a relationship with days to report, as younger victims more often take longer to report, specifically victims under the age of 20.

```python
ax = crimes_df.plot.scatter(x='Crime Code', y='Victim Age', c='Days To Report', alpha=0.5)
ax;
```

```python
crimes_df.plot.scatter(x='Crime Code', y='Days To Report', alpha=0.5, c='darkolivegreen');
```

```python
crimes_df.groupby('Crime Code')['Days To Report'].mean().plot.line();
```
